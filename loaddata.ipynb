```python
def select_columns(data):
    df = json.loads(data)
    return{
        "book_id": df["book_id"],
        "title": df["title"],
        "ratings_count": df["ratings_count"],
        "authors": df["authors"],
        "average_rating": df["average_rating"],
        "description": df["description"],
        "image_url": df["image_url"]
        
    }
```

```python
books = []

with gzip.open("C:/Users/DELL/Downloads/goodreads_books.json.gz", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        cols = select_columns(line)
        try:
            ratings = int(cols["ratings_count"])
        except ValueError:
            continue
        if ratings > 200:
            books.append(cols)
        
```

```python
booksdf = pd.DataFrame.from_dict(books)
```

```python
def select_columns_authors(data):
    df = json.loads(data)
    return{
        "author_id": df["author_id"],
        "name": df["name"],
        "ratings_count": df["ratings_count"]
        
    }
```

```python
authors = []

with gzip.open("C:/Users/DELL/Downloads/goodreads_book_authors.json.gz", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        cols = select_columns_authors(line)
        try:
            ratings = int(cols["ratings_count"])
        except ValueError:
            continue
        if ratings > 20:
            authors.append(cols)
```
