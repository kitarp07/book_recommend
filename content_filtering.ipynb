```python
new_df = data[["book_id", "title","description","genres", "tags", "image_url", "author_name"]]
new_df
```

```python
new_df['tags'] =new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
dff = new_df.head(1000)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000, stop_words = 'english')
```

```python
book_vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(book_vectors)
```

```python
similarity[0]
```

```python
new_df
```

```python
def recommend_book(book):
    book_index = new_df[new_df['title']==book].index[0]
    similarity_array = similarity[book_index ]
    recommended_books = sorted(list(enumerate(similarity_array)), reverse=True, key=lambda x: x[1])[1:10]
    
    books = []
    for i in recommended_books:
        print(i)
        items = []
        temp_df = (new_df.iloc[i[0]])
        items.append(temp_df['title'])
        items.append(temp_df['author_name'])
        items.append(temp_df['image_url'])
       

        books.append(items)
    return books
    
```

```python
recommend_book("Harry Potter and the Sorcerer's Stone (Harry Potter, #1)")
```
