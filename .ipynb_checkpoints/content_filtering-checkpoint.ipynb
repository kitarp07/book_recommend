```python
data = data.loc[data['author_name'].apply(lambda x: not isinstance(x, float))]
data
```

```python
data['author_name'] = data['author_name'].apply(lambda x: x.replace(" ", ""))
data
```

```python
d = data.head(5000)
d
```

```python
d['description'] = d['description'].astype(str)
d['genres'] = d['genres'].astype(str)
d['author_name'] = d['author_name'].astype(str)
d
```

```python
d['description']=d['description'].apply(lambda x: x.replace("]", ""))
d['description']=d['description'].apply(lambda x: x.replace("[", ""))
d['description']=d['description'].apply(lambda x: x.replace(".", ""))
d['description']=d['description'].apply(lambda x: x.replace("'", ""))
d['description'][0]
```

```python
d['genres']=d['genres'].apply(lambda x: x.replace("[", ""))
d['genres']=d['genres'].apply(lambda x: x.replace("]", ""))
d['genres']=d['genres'].apply(lambda x: x.replace("'", ""))
d['genres'][0]
```

```python
d['tags'] = d['description'] +"," + d['genres'] + ", " + d["author_name"]
d['tags'][0]
```

```python
d['tags'][2]
```

```python
new_df = d[["book_id", "title", "tags"]]
new_df
```

```python
new_df['tags'] =new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
dff = new_df.head(1000)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=500, stop_words = 'english')
```

```python
book_vectors = cv.fit_transform(dff['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
len(cv.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(book_vectors)
```

```python
similarity[0]
```

```python
dff
     
```

```python
def recommend_book(book):
    book_index = dff[dff['title']==book].index[0]
    similarity_array = similarity[book_index ]
    recommended_books = sorted(list(enumerate(similarity_array)), reverse=True, key=lambda x: x[1])[1:10]
    
    for i in recommended_books:
        print (dff.iloc[i[0]].title)
    
```

```python
recommend_book('Search the Seven Hills')
```
