```python
def select_columns(data):
    df = json.loads(data)
    return{
        "book_id": df["book_id"],
        "title": df["title"],
        "ratings_count": df["ratings_count"],
        "authors": df["authors"],
        "average_rating": df["average_rating"],
        "description": df["description"],
        "image_url": df["image_url"]
        
    }
```

```python
books = []

with gzip.open("goodreads_books.json.gz", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        cols = select_columns(line)
        try:
            ratings = int(cols["ratings_count"])
        except ValueError:
            continue
        if ratings > 20:
            books.append(cols)
        
```

```python
books[0]
```

```python
booksdf = pd.DataFrame.from_dict(books)
```

```python
booksdf.isnull().sum()
```

```python
def select_columns_authors(data):
    df = json.loads(data)
    return{
        "author_id": df["author_id"],
        "name": df["name"],
        "ratings_count": df["ratings_count"]
        
    }
```

```python
authors = []

with gzip.open("goodreads_book_authors.json.gz", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        cols = select_columns_authors(line)
        try:
            ratings = int(cols["ratings_count"])
        except ValueError:
            continue
        if ratings > 20:
            authors.append(cols)
```

```python
authorsdf = pd.DataFrame.from_dict(authors)
```

```python
import gzip

#open file in read mode in streaming fashion

with gzip.open("goodreads_book_genres_initial.json.gz", "r") as f:
    l2 = f.readline()
l2
```

```python
def select_columns_genres(data):
    df = json.loads(data)
    return{
        "book_id": df["book_id"],
        "genres": df["genres"]
        
    }
```

```python
genres = []

with gzip.open("goodreads_book_genres_initial.json.gz", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        cols = select_columns_genres(line)
        genres.append(cols)
```

```python
genres
```

```python
genresdf = pd.DataFrame.from_dict(genres)
```

```python
genresdf['genres'] = genresdf['genres'].apply(lambda x: list(x.keys()))
```

```python
genresdf.to_csv("C:/Users/user/Desktop/bookrecommend/genres.csv")
```
